```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv("/content/Salary_Data.csv")
```

```python
data_x_train=df.loc[:19,["YearsExperience"]]
data_y_train=df.loc[:19,["Salary"]]
data_x_test=df.loc[20:30,["YearsExperience"]]
data_y_test=df.loc[20:30,["Salary"]]
```

```python
model = linear_model.LinearRegression()
model.fit(data_x_train,data_y_train)
data_y_prediction=model.predict(data_x_test)
print("The Mean Squared error :",mean_squared_error(y_true=data_y_test,y_pred=data_y_prediction))
print("The Wights :",model.coef_)
print("Intersection :",model.intercept_)
```

```python
print("Accurecy Of This Model Is : "+str(r2_score(data_y_test,data_y_prediction)*100)+"%")
plt.scatter(data_x_test,data_y_test)
plt.plot(data_x_test,data_y_prediction)
plt.show()
```
